```python
import matplotlib.pyplot as plt
import numpy as np

# Data for dynamic range quantization
quantized_data = {
    "cpu 4 threads (xnnpack)": {
        "avg_time": 103222,
        "std_dev": 7300
    },
    "cpu 2 threads": {
        "avg_time": 107997,
        "std_dev": 2109
    },
    "cpu 1 threads": {
        "avg_time": 111545,
        "std_dev": 1930
    },
    "cpu 2 threads (xnnpack)": {
        "avg_time": 112655,
        "std_dev": 1723
    },
    "cpu 1 threads (xnnpack)": {
        "avg_time": 114236,
        "std_dev": 1812
    },
    "gpu-default": {
        "avg_time": 124948,
        "std_dev": 2816
    },
    "cpu 4 threads": {
        "avg_time": 127433,
        "std_dev": 8085
    }
}

# Data for unquantized model
unquantized_data = {
    "cpu 4 threads": {
        "avg_time": 32342.5,
        "std_dev": 4581
    },
    "cpu 4 threads (xnnpack)": {
        "avg_time": 36086,
        "std_dev": 3203
    },
    "cpu 2 threads": {
        "avg_time": 39234.3,
        "std_dev": 818
    },
    "cpu 2 threads (xnnpack)": {
        "avg_time": 45355.6,
        "std_dev": 13556
    },
    "cpu 1 threads (xnnpack)": {
        "avg_time": 62408.6,
        "std_dev": 1483
    },
    "cpu 1 threads": {
        "avg_time": 66269.7,
        "std_dev": 2126
    },
    "gpu-default": {
        "avg_time": 66347.7,
        "std_dev": 2466
    }
}

def calculate_reduction(quantized_data, unquantized_data):
    reduction_data = {}
    for config in quantized_data:
        reduction = quantized_data[config]["avg_time"] - unquantized_data[config]["avg_time"]
        reduction_data[config] = reduction
    return reduction_data

def plot_data(quantized_data, unquantized_data, title):
    configs = list(quantized_data.keys())
    avg_times1 = [quantized_data[config]["avg_time"] for config in configs]
    avg_times2 = [unquantized_data[config]["avg_time"] for config in configs]
    std_devs1 = [quantized_data[config]["std_dev"] for config in configs]
    std_devs2 = [unquantized_data[config]["std_dev"] for config in configs]
    
    reduction_data = calculate_reduction(quantized_data, unquantized_data)
    reductions = [reduction_data[config] for config in configs]

    bar_width = 0.35
    index = np.arange(len(configs))

    plt.figure(figsize=(12, 6))
    plt.bar(index, avg_times1, bar_width, label='Quantized', color='skyblue', yerr=std_devs1, capsize=5)
    plt.bar(index + bar_width, avg_times2, bar_width, label='Unquantized', color='salmon', yerr=std_devs2, capsize=5)
    plt.xlabel('Configuration')
    plt.ylabel('Average Inference Time (microseconds)')
    plt.title(title)
    plt.xticks(index + bar_width / 2, configs, rotation=45, ha='right')
    plt.legend()
    for i in range(len(configs)):
        plt.text(index[i] + bar_width / 2, max(avg_times1[i], avg_times2[i]), f'{reductions[i]:.2f}', ha='center', va='bottom', color='black')
    plt.tight_layout()
    plt.show()

plot_data(quantized_data, unquantized_data, 'Comparison of Quantized and Unquantized Models')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for unquantized model
unquantized_avg_times = {
    "cpu 4 threads": 33904.9,
    "cpu 4 threads (xnnpack)": 39220.2,
    "cpu 2 threads (xnnpack)": 41921.3,
    "cpu 2 threads": 44573,
    "gpu-default": 57705.5,
    "cpu 1 threads (xnnpack)": 61070,
    "cpu 1 threads": 62723.4
}

# Data for FP16 quantization
fp16_avg_times = {
    "cpu 4 threads (xnnpack)": 34673.5,
    "cpu 4 threads": 39937.7,
    "cpu 2 threads": 48602.2,
    "cpu 2 threads (xnnpack)": 55170.9,
    "cpu 1 threads (xnnpack)": 62946,
    "gpu-default": 64018.8,
    "cpu 1 threads": 64855.8
}

# Data for dynamic range quantization
dynamic_range_avg_times = {
    "cpu 1 threads (xnnpack)": 98936.8,
    "cpu 4 threads (xnnpack)": 107331,
    "cpu 2 threads (xnnpack)": 107661,
    "cpu 2 threads": 115431,
    "cpu 1 threads": 121102,
    "gpu-default": 121437,
    "cpu 4 threads": 128087
}

def calculate_speedup_factor(unquantized_data, quantized_data):
    speedup_factors = {}
    for config, unquantized_time in unquantized_data.items():
        quantized_time = quantized_data.get(config)
        if quantized_time is not None:
            speedup_factors[config] = quantized_time / unquantized_time
    return speedup_factors

def plot_speedup_factor(speedup_factors):
    configs = list(speedup_factors.keys())
    speedups = list(speedup_factors.values())

    plt.figure(figsize=(10, 6))
    plt.plot(configs, speedups, marker='o', linestyle='-')
    plt.xlabel('Configuration')
    plt.ylabel('Speedup Factor')
    plt.title('Speedup Factor of Unquantized Model vs. Quantized Versions')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

#  speedup factors
fp16_speedup_factors = calculate_speedup_factor(unquantized_avg_times, fp16_avg_times)
int8_speedup_factors = calculate_speedup_factor(unquantized_avg_times, dynamic_range_avg_times)


plot_speedup_factor(fp16_speedup_factors)
plot_speedup_factor(int8_speedup_factors)
```
